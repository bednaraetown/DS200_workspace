### Step 8. Suppose you are in a country that people can legally start to drink alcohol when they are older than 16 (not including 16). Then create a new column for the DataFrame called `legal_drinker` by applying the `check_drinker` function to every students on the DataFrame. Show the first 5 rows of the resulting DataFrame.

```python
# Create the drinker function here
def check_drinker(x):
    if x > 16:
        return True
    else:
        return False


```
